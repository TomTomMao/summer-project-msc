```python
import pandas as pd
import jellyfish
```

```python
df = pd.read_csv('./transaction_cleanedtest copy.csv')
```

```python
# get unique Transaction Description
unique_transaction_description = df['Transaction Description'].unique()
```

```python
unique_transaction_description
unique_transaction_description_words = [x.split() for x in unique_transaction_description]
unique_transaction_description_words_flatten = [item for sublist in unique_transaction_description_words for item in sublist]
```

```python
top10casinos  = [
    "Hippodrome",
    "Genting",
    "Victoria Gate",
    "Alea",
    "Aspers",
    "Forty Five",
    "Park Lane Club",
    "Horizons",
    "Genting",
    "Admiral"
]

casinos = top10casinos
```

```python
def show_matches(casinos, descriptions, distance_function):
    # return a sorted list of tuples containing the casino, description and the distance between them
    # if the distance is greater than the similarity threshold
    matches = []
    for casino in casinos:
        for description in descriptions:
            matches.append((casino, description, distance_function(casino.lower(), description.lower())))
    return sorted(matches, key=lambda x: x[2], reverse=True)
```

```python
matches = show_matches(casinos, unique_transaction_description, jellyfish.levenshtein_distance)
sorted_matches = sorted(matches, key=lambda x: x[2], reverse=False)
sorted_matches
```

```python
matches = show_matches(casinos, unique_transaction_description, jellyfish.hamming_distance)
sorted_matches = sorted(matches, key=lambda x: x[2], reverse=False)
sorted_matches
```

```python
matches = show_matches(casinos, unique_transaction_description, jellyfish.jaro_winkler_similarity)
sorted_matches = sorted(matches, key=lambda x: x[2], reverse=True)
sorted_matches
```

```python
matches = show_matches(casinos, unique_transaction_description_words_flatten, jellyfish.jaro_winkler_similarity)
sorted_matches = sorted(matches, key=lambda x: x[2], reverse=True)
sorted_matches
```
