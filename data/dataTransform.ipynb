```python
def checkCityCountry():
    # create a city country map
    cityCountryMap = {}
    cities = list(df["Location City"])
    countries = list(df["Location Country"])
    assert(len(cities) == len(countries))
    uniquePairs = list(set(zip(cities, countries)))
    # check if there exists dupilicate cities in the unqiuePairs
    citiesFromPairs = [city for city,_ in uniquePairs]
    duplicatedCities = []
    uniqueCities = []
    for city in citiesFromPairs:
        if city in uniqueCities:
            duplicatedCities.append(city)
        else:
            uniqueCities.append(city)
    print("duplicatedCities:", duplicatedCities)
    return ([(city, country) for city, country in uniquePairs if city in duplicatedCities])
checkCityCountry()
```

### convert to wide version

```python
AMOUNT_TYPES = ["debit", "credit"]
TRANSACTION_TYPES = ["Code","BGC","BP","CPT","CHQ","DD","DEB","DEP","FEE","FPI","FPO","PAY","TFR","SO","None"]
AMOUNT_TYPES_TRANSACTION_TYPES = []
def createCategory(amountType, transactionType):
    if transactionType not in TRANSACTION_TYPES[:-1]:
        transactionType = "None"
    return amountType + "-" + transactionType
for amountType in AMOUNT_TYPES:
    for transactionType in TRANSACTION_TYPES:
        AMOUNT_TYPES_TRANSACTION_TYPES.append(createCategory(amountType, transactionType))
```

```python
AMOUNT_TYPES_TRANSACTION_TYPES
```

```python
df3["time"] = pd.to_datetime(df3["TransactionDate"])
df3.sort_values(by = ["time", "AmountType", "TransactionType"])
```

```python
df3.notnull().sum()
```

```python
def createEmptyDateData(date:str) -> dict:
    data = {"date": date}
    for category in AMOUNT_TYPES_TRANSACTION_TYPES:
        data[category] = {"totalTransactionAmount":0, "transactionCount":0}
    return data
dateDatabase = {}
for label, (date, amountType, transactionType, totalTransactionAmount, transactionCount) in df3.iterrows():
    category = createCategory(amountType, transactionType)
    # print(date, category, totalTransactionAmount, transactionCount)
    if date not in dateDatabase:
        dateDatabase[date] = createEmptyDateData(date)
    dateDatabase[date][category]["totalTransactionAmount"] = totalTransactionAmount
    dateDatabase[date][category]["transactionCount"] = transactionCount
assert len(dateDatabase.keys()) == len(set(df3["TransactionDate"]))
```

```python
transactionAmountDatabase = []
transactionCountDatabase = []
for key,value in dateDatabase.items():
    amountData = {'datestr': value["date"]}
    countData = {'datestr': value["date"]}
    for category in AMOUNT_TYPES_TRANSACTION_TYPES:
        amountData[category] = value[category]["totalTransactionAmount"]
        countData[category] = value[category]["transactionCount"]
    transactionAmountDatabase.append(amountData)
    transactionCountDatabase.append(countData)
print(transactionAmountDatabase[0:5])
print(transactionCountDatabase[0:5])
```

```python
# convert data to data frame
df3_wide_amount = pd.DataFrame(transactionAmountDatabase)
df3_wide_count = pd.DataFrame(transactionCountDatabase)

# convert date string to datetype
df3_wide_amount["date"] = pd.to_datetime(df3_wide_amount["datestr"], dayfirst=True)
df3_wide_count["date2"] = pd.to_datetime(df3_wide_count["datestr"], dayfirst=True)

# add suffix after the column names
amountColumnNameMapping = {oldColumnName: oldColumnName + "-" + "amount" for oldColumnName in list(df3_wide_amount.columns) if oldColumnName != "date"}
countColumnNameMapping = {oldColumnName: oldColumnName + "-" + "count" for oldColumnName in list(df3_wide_count.columns) if oldColumnName != "date"}
df3_wide_amount = df3_wide_amount.rename(amountColumnNameMapping, axis=1)
df3_wide_count = df3_wide_count.rename(countColumnNameMapping, axis=1)

# concat columns together
df4 = pd.concat([df3_wide_amount, df3_wide_count], axis = 1)
df4 = df4.drop(["datestr-count","datestr-amount", "date2-count"], axis = 1)
df4
```

```python
# Hakuna-Patata. (2022, February 16). Answer to ‘Add missing dates to pandas dataframe’. Stack Overflow. https://stackoverflow.com/a/71136824

# first create a DataFrame with all dates between specified start<-->end using pd.date_range()
all_dates = pd.DataFrame(pd.date_range(df4['date'].min(), df4['date'].max()), columns=['date'])

# from the all_dates DataFrame, left join onto the DataFrame with missing dates
df4 = all_dates.merge(right=df4, how='left', on='date')
df4 = df4.fillna(0)
```

```python
# Add Recency
def get_recency_map(_df: pd.DataFrame):
    '''
        df: a dataframe with 'Transaction Date' (pd.Timestamp) and 'Transaction Description' (string) column
        Time complexity: O(n),
        Space complexity: O(k), k = number of the set of the transaction descriptions
        return a dictionary (transaction Description -> Recency), recency value is the number of dates between 
        the latest transaction with the same transaction description to today.
    '''

    # iterate through the rows,
        # if transactiondescription is empty: next iteration
        # elif transaction date is greater than today, raise an error.
        # elif transaction description not in lastestDate, add it in to the latestDate
        # elif transaction description is in lastestDate:
            # update latestDate[description] if the new transaction description is later than the exist one in the latestDate
    latestDate = {} # transaction description->latestDate
    for index, row in _df.iterrows():
        transactionDescription: str = row['Transaction Description']
        transactionDate: pd.Timestamp = row['Transaction Date']

        if transactionDescription == "":
            continue
        elif transactionDate > pd.Timestamp.today():
            raise ValueError(f"transaction Date must be earlier than today, index: {index}, transactionDate: {transactionDate}")
        elif transactionDescription not in latestDate:
            latestDate[transactionDescription] = transactionDate
        elif latestDate[transactionDescription] < transactionDate:
            latestDate[transactionDescription] = transactionDate
    
    return {transactionDescription: (pd.Timestamp.today() - transactionDate).days for transactionDescription, transactionDate in latestDate.items()}
get_recency_map(df)
```

```python
def get_frequency(_df: pd.DataFrame, frequencyType: str = 'total'):
    '''
        
    '''
    if frequencyType=='total':
        transactionCount = {}
        for index, row in _df.iterrows():
            transactionCount[row['Transaction Description']] = transactionCount.get(row['Transaction Description'], 0) + 1
        return transactionCount
    
    if frequencyType=='avgYear':
        transactionFrequency = {}

        # add: 'transaction description' -> [transaction date] into transactionDates
        for index, row in _df.iterrows():
            transactionDescription = row['Transaction Description']
            transactionDate = row['Transaction Date']
            if transactionDescription not in transactionFrequency:
                transactionFrequency[transactionDescription] = {str(transactionDate.year): 1}
            else:
                transactionFrequency[transactionDescription][str(transactionDate.year)] = transactionFrequency[transactionDescription].get(str(transactionDate.year), 0) + 1
        return transactionFrequency


get_frequency(df, 'avgYear')
    
```
